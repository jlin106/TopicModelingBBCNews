```python
# Creating a topic2wordproportions dataframe and csv file containing a dataframe of 
# the topic proportions, tag and tag ID for each document
import gensim
from gensim import corpora
import sqlite3
import sys
import csv
import re
import itertools
import pandas as pd
```

```python
model =  gensim.models.LdaModel.load('lda_model175')
```

```python
# recreate the corpus
text_lemmas = []
con = sqlite3.connect('news.db')
cur = con.cursor()
for row in range(36162):
	cur.execute("SELECT Words FROM BBCnews WHERE ID = ?;", (row + 1,))
	rowText = cur.fetchone()[0]
	rowText = re.sub(r'\s', '', rowText).split(',')
	text_lemmas.append(rowText)
id2word = corpora.Dictionary(text_lemmas)
id2word.filter_extremes(no_below = 30)
id2word.filter_tokens(bad_ids=[id2word.token2id['v'], id2word.token2id['p'],id2word.token2id['x'], id2word.token2id['b'], id2word.token2id['c'], id2word.token2id['q'],id2word.token2id['e'], id2word.token2id['j'],id2word.token2id['r'], id2word.token2id['g'], id2word.token2id['l'], id2word.token2id['n'], id2word.token2id['f'], id2word.token2id['w'],id2word.token2id['h'], id2word.token2id['mp'], id2word.token2id['de'], id2word.token2id['ap'],id2word.token2id['el'], id2word.token2id['le'], id2word.token2id['1m']])
id2word.compactify()
corpus = [id2word.doc2bow(text) for text in text_lemmas]
```

```python
# create a dataframe for the topic proportions for each document
topics = [str(x) for x in range(175)]
topicID = ["T" + topic for topic in topics]
docs = [str(x) for x in range(36162)]
```

```python
# setting the dimensions and filling with default 0 value
topic_pro_df = pd.DataFrame(index = docs, columns = topicID)
topic_pro_df = topic_pro_df.fillna(0)
topic_pro_df
```

```python
# updating the dataframe with the topic proportions
for i in range(36162):
    list_pro = model.get_document_topics(corpus[i])
    for pair in list_pro:
        col = 'T'+str(pair[0])
        topic_pro_df.iloc[i, topic_pro_df.columns.get_loc(col)] = pair[1]
```

```python
# save the dataframe as a csv file
topic_pro_df.to_csv('topic2wordproportions.csv')
```

```python
# converting the csv to a sqlite database topics.db
con = sqlite3.connect('topics.db')
cur = con.cursor()
topic_pro_df.to_sql('topic2wordpp_df',con, index_label='ID')
con.commit()
con.close()
```

```python
# taking the tag IDs from the BBCnews table
tagID = []
con = sqlite3.connect('news.db')
cur = con.cursor()
cur.execute("SELECT TagID FROM BBCnews;")
for i in cur:
    tagID.append((i[0]))
```

```python
# creating and adding column for tag ID in new topic2wordpp_df table
con = sqlite3.connect('topics.db')
cur = con.cursor()
cur.execute("ALTER TABLE topic2wordpp_df ADD TagID Text;")
index = 0
for i in tagID:
    cur.execute("UPDATE topic2wordpp_df SET TagID = ? WHERE ID = ?;", (i,index,))
    index += 1
con.commit()
con.close()
```
