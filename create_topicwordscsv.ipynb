```python
import gensim
from gensim import corpora
import sqlite3
import sys
import csv
import re
import pandas as pd
```

```python
# import lda model
model =  gensim.models.LdaModel.load('lda_model175')
```

```python
topics = [str(x) for x in range(175)]
term_ids = ['Term0', 'Term1', 'Term2', 'Term3', 'Term4', 'Term5', 'Term6', 'Term7', 'Term8', 'Term9']
onlyterms_df = pd.DataFrame(index = topics, columns = term_ids)
```

```python
for x in range(175):
    terms = re.sub('[0123456789.*""]', '', model.print_topic(x)).split(' + ')
    for y in range(10):
        col = 'Term'+str(y)
        onlyterms_df.iloc[x, onlyterms_df.columns.get_loc(col)] = terms[y]
onlyterms_df
```

```python
# save the dataframe as a csv file
onlyterms_df.to_csv('onlyterms_df.csv')
```

```python
# create pandas df with the topics and the words
topicwords_df = pd.DataFrame(columns = ['Topic ID', 'Words'])
for x in range(175):
    idcol = 'T'+str(x)
    topicwords_df = topicwords_df.append([{'Topic ID':idcol, 'Words':re.sub('[0123456789.*]', '', model.print_topic(x))}],ignore_index=True)
topicwords_df = topicwords_df.set_index('Topic ID')
topicwords_df
```

```python
# save the dataframe as a csv file
topicwords_df.to_csv('topicwords_df.csv')
```
