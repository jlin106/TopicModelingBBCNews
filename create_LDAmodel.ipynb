```python
import gensim
from gensim import corpora
from gensim.models import CoherenceModel
import sqlite3
import sys
import csv
import re
import pyLDAvis
import pyLDAvis.gensim
import itertools
```

```python
text_lemmas = []

# number of documents
num_rows = 36162

con = sqlite3.connect('news.db')
cur = con.cursor()

# extract the words for each document from the database
for row in range(num_rows):
	cur.execute("SELECT Words FROM BBCnews WHERE ID = ?;", (row + 1,))
	rowText = cur.fetchone()[0]
	rowText = re.sub(r'\s', '', rowText).split(',')
	text_lemmas.append(rowText)
```

```python
# assign number ID to each word
id2word = corpora.Dictionary(text_lemmas)
merged = [list(itertools.chain(*text_lemmas))]
```

```python
# filter the tokens 
id2word.filter_extremes(no_below = 30)
id2word.filter_tokens(bad_ids=[id2word.token2id['v'], id2word.token2id['p'], id2word.token2id['x'], id2word.token2id['b'], id2word.token2id['c'], id2word.token2id['q'], id2word.token2id['e'], id2word.token2id['j'], id2word.token2id['r'], id2word.token2id['g'], id2word.token2id['l'], id2word.token2id['n'], id2word.token2id['f'], id2word.token2id['w'], id2word.token2id['h']])
```

```python
# compact the new dictionary after filtering
id2word.compactify()
print(id2word)
```

```python
# create a corpus
corpus = [id2word.doc2bow(text) for text in text_lemmas]
```

```python
# print the words
corpus_merged = [id2word.doc2bow(text) for text in merged]
word_list = corpus_merged[0]
word_list.sort(key=lambda tup: tup[0])
for tup in word_list:
    print(id2word[tup[0]])
```

```python
# create the lda model with 175 topics
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=175, 
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
for num in range(0, 175):
    print(num, ',\n', lda_model.print_topic(num))
```

```python
# visualization
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, id2word)
vis
```

```python
# save the lda model
lda_model.save('lda_model175')
```
